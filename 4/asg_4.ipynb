```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yaml
```

```python
with open("config.yml") as config:
    input_files = yaml.safe_load(config)
    df = pd.read_csv(input_files["price"])

plt.scatter(df["size"],df["price"],color = "red")
plt.title("scatter of price increasement")
plt.xlabel("size (m2)")
plt.ylabel("ekkies")
plt.grid()
plt.show()
```

```python
def compute_cost(X,y,theta):
    n = len(X)
    m = len(y)
    x0 = np.ones((n,1))
    X = np.concatenate((x0,X.values.reshape(-1,1)),axis=1)
    fractor = 1/(2*m)
    h = np.dot(X,theta)

    cost = np.sum((h-y)**2)
    pred_cost = fractor*cost

    return pred_cost
```

```python
# plotting
# Example usage
X = df["size"]# Assuming two input features
y = df["price"]  # True output values
theta = np.array([0,0])  # Initial parameter values

# Calculate the cost
cost = compute_cost(X, y, theta)
print("Cost:", cost)
```

```python
def gradient_descent(X,alpha,y,theta,num_iters):
    m = len(y)
    n = len(X)
    x0 = np.ones((n,1))
    X = np.concatenate((x0,X.values.reshape(-1,1)),axis=1)

    fractor = alpha *(1/m)
    h = np.dot(X,theta)
    cost = np.sum(np.dot(h-y,X))
    pred_cost = fractor*cost
    theta = theta - pred_cost
    return pred_cost, theta
```

```python
# plotting
# Example usage
X = df["size"]# Assuming two input features
y = df["price"]  # True output values
theta = np.array([0,0])  # Initial parameter values

# Calculate the cost
cost,thata = gradient_descent(X,-5,y,theta,1)
print("Cost:", cost , "Theta:",thata)
```
